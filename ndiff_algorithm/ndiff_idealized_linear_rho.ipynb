```python
from netCDF4 import Dataset
import numpy as np
import matplotlib.pyplot as plt
```

```python
datapath = '/home/ashao/git/ashao_forks/MOM6-examples/ocean_only/tracer_mixing/rho/'
prog_vars = Dataset(datapath + 'prog.nc').variables
init_vars = Dataset(datapath + 'Initial_state.nc').variables
```

```python
h = prog_vars['h'][:,:,0,:]
temp = prog_vars['temp'][:,:,0,:]
salt = prog_vars['salt'][:,:,0,:]
depth = h.cumsum(axis = 1) - 0.5*h

h0 = init_vars['h'][0,:,0,:]
temp0 = init_vars['Temp'][0,:,0,:]
salt0 = init_vars['Salt'][0,:,0,:]
depth0 = h0.cumsum(axis=0) - 0.5*h0

xh = prog_vars['xh'][:]
xh = np.tile(xh.transpose(),(depth.shape[1],1))
xh.shape
```

```python
temp0.max()
```

```python
S_levels = np.linspace(salt0.min(),salt0.max(),10)
T_levels = np.linspace(temp0.min(),temp0.max(),10)
plt.contour(xh,depth0,salt0,S_levels,colors='black')
plt.contour(xh,depth0,temp0,T_levels,colors='red')
plt.pcolormesh(xh,depth0,salt0)
plt.gca().invert_yaxis()
plt.show()
```

```python
S_levels = np.linspace(salt[0,:,:].min(),salt.max(),10)
T_levels = np.linspace(temp[0,:,:].min(),temp.max(),10)
plt.contour(xh,depth[-1,:,:],salt[-1,:,:],S_levels,colors='black')
plt.contour(xh,depth[-1,:,:],temp[-1,:,:],T_levels,colors='red')
plt.gca().invert_yaxis()
plt.show()
```

```python
plt.pcolormesh(xh,depth[-1,:,:],temp[-1,:,:])
plt.gca().invert_yaxis()
plt.show()
```
